#### **Helper Functions**

###### **Weighted AKME Function**


```r
akme_survival <- function(dat,
                          z_value,
                          time_var = "T",
                          event_var = "delta",
                          treat_var = "Z",
                          weight_var = "w") {

  dsub <- dat[dat[[treat_var]] == z_value, , drop = FALSE]

  t <- dsub[[time_var]]
  d <- dsub[[event_var]]
  w <- dsub[[weight_var]]

  # event times only
  event_times <- sort(unique(t[d == 1]))

  if (length(event_times) == 0) {
    return(data.frame(time = 0, surv = 1))
  }

  surv <- numeric(length(event_times))
  s_now <- 1

  for (j in seq_along(event_times)) {
    tj <- event_times[j]

    # weighted risk set at tj
    Yw <- sum(w[t >= tj], na.rm = TRUE)

    # weighted events at tj
    dw <- sum(w[t == tj & d == 1], na.rm = TRUE)

    if (Yw > 0) {
      s_now <- s_now * (1 - dw / Yw)
    }

    surv[j] <- s_now
  }

  out <- data.frame(
    time = c(0, event_times),
    surv = c(1, surv)
  )

  rownames(out) <- NULL
  out
}
```

###### **Extract Survival quantile**

```r
surv_quantile <- function(S_df, q) {
  target <- 1 - q
  idx <- which(S_df$surv <= target)[1]

  if (is.na(idx)) return(NA_real_)
  S_df$time[idx]
}
```

##### **Function to compute PSQE from AKME**

```r
compute_psqe_from_akme <- function(dat, weight_var = "w") {

  S1 <- akme_survival(dat, z_value = 1, weight_var = weight_var)
  S0 <- akme_survival(dat, z_value = 0, weight_var = weight_var)

  q_levels <- c(0.25, 0.50, 0.75)

  Q1 <- sapply(q_levels, function(q) surv_quantile(S1, q))
  Q0 <- sapply(q_levels, function(q) surv_quantile(S0, q))

  out <- data.frame(
    quantile = q_levels,
    Q1 = Q1,
    Q0 = Q0,
    PSQE = Q1 - Q0
  )

  list(
    S1 = S1,
    S0 = S0,
    psqe = out
  )
}
```

##### **Function for MEthod I code**

```r
akme_method1 <- function(dat) {
  dat$w <- dat$w_ps
  compute_psqe_from_akme(dat, weight_var = "w")
}
```

##### **Function for MEthod II code**

```r
akme_method2 <- function(dat) {
  dat$w <- dat$W_final
  compute_psqe_from_akme(dat, weight_var = "w")
}
```

##### **Function for MEthod II code**

```r
akme_method3 <- function(dat,
                         ps_formula = Z ~ X1 + X2 + X3 + X4 + X5 + X_61 + X_62) {

  # Step 1: survey-weighted PS model
  design_ps <- svydesign(
    ids = ~cluster,
    strata = ~stratum,
    weights = ~w_svy,
    data = dat,
    nest = TRUE
  )

  ps_fit <- svyglm(
    ps_formula,
    design = design_ps,
    family = quasibinomial()
  )

  ps_hat <- predict(ps_fit, type = "response")

  # avoid 0 and 1
  eps <- 1e-6
  ps_hat <- pmin(pmax(ps_hat, eps), 1 - eps)

  # Step 2: IPTW
  w_ps_m3 <- ifelse(dat$Z == 1, 1 / ps_hat, 1 / (1 - ps_hat))

  # Step 3: final combined weight
  dat$w <- dat$w_svy * w_ps_m3

  compute_psqe_from_akme(dat, weight_var = "w")
}
```

```r
print_psqe_results <- function(res_M1, res_M2, res_M3) {
  cat("\nMethod I PSQE\n")
  print(res_M1$psqe)

  cat("\nMethod II PSQE\n")
  print(res_M2$psqe)

  cat("\nMethod III PSQE\n")
  print(res_M3$psqe)
}
```

```r
run_one_psqe <- function(dat) {
  list(
    M1 = akme_method1(dat)$psqe,
    M2 = akme_method2(dat)$psqe,
    M3 = akme_method3(dat)$psqe
  )
}
```

```r
run_psqe_batch <- function(all_dat, start = 1, end = length(all_dat)) {

  end <- min(end, length(all_dat))

  results <- vector("list", end - start + 1)

  for (i in start:end) {
    dat <- all_dat[[i]]

    results[[i - start + 1]] <- tryCatch(
      run_one_psqe(dat),
      error = function(e) {
        message(sprintf("Error at dataset %d: %s", i, e$message))
        NULL
      }
    )
  }

  names(results) <- paste0("rep_", start:end)
  results
}
```

```r
res_to_df <- function(res_obj) {
  do.call(rbind, lapply(names(res_obj), function(rep_name) {
    rep_res <- res_obj[[rep_name]]

    do.call(rbind, lapply(names(rep_res), function(method_name) {
      df <- rep_res[[method_name]]
      df$rep <- rep_name
      df$method <- method_name
      df
    }))
  }))
}
```
